| **Phase 1** | Contrôle Qualité des Données (Data QC) |

## 📖 Dictionnaire des Variables — UCI Credit Card Default

### Informations générales
| Variable | Type | Description |
|---|---|---|
| `ID` | Entier | Identifiant unique du client (1 à 30 000) — non utilisé en modélisation |
| `DEFAULT` | Binaire | **Variable cible** : `1` = défaut de paiement le mois suivant, `0` = pas de défaut |

---

### Profil du client
| Variable | Type | Modalités / Plage | Description |
|---|---|---|---|
| `LIMIT_BAL` | Continu | [10 000 ; 1 000 000 NT$] | Montant de la ligne de crédit accordée au client (en dollars taïwanais) |
| `SEX` | Catégoriel | `1` = Homme, `2` = Femme | Genre du client |
| `EDUCATION` | Catégoriel | `1` = études supérieures, `2` = université, `3` = lycée, `4` = autre, `0/5/6` = non documenté | Niveau d'éducation du client |
| `MARRIAGE` | Catégoriel | `1` = marié(e), `2` = célibataire, `3` = autre, `0` = non documenté | Statut matrimonial du client |
| `AGE` | Entier | [21 ; 79 ans] | Âge du client en années |

---

### Historique des retards de paiement (avril à septembre 2005)
> **Echelle de retard** : `-2` = pas de consommation, `-1` = payé en avance,
> `0` = paiement ponctuel, `1` = retard 1 mois, ..., `8` = retard 8 mois ou plus

| Variable | Mois correspondant | Description |
|---|---|---|
| `PAY_0` | Septembre 2005 | Statut de remboursement le mois le plus récent |
| `PAY_2` | Août 2005 | Statut de remboursement — 2 mois avant |
| `PAY_3` | Juillet 2005 | Statut de remboursement — 3 mois avant |
| `PAY_4` | Juin 2005 | Statut de remboursement — 4 mois avant |
| `PAY_5` | Mai 2005 | Statut de remboursement — 5 mois avant |
| `PAY_6` | Avril 2005 | Statut de remboursement — 6 mois avant |

> ⚠️ **Note** : il n'existe pas de `PAY_1` dans le dataset original —
> c'est une numérotation de la source UCI, non une erreur.

---

### Montant des relevés de facturation (avril à septembre 2005)
> Montant dû sur le relevé mensuel de carte de crédit, en NT$.
> Une valeur **négative** indique un crédit excédentaire
> (le client a trop remboursé le mois précédent).

| Variable | Mois correspondant |
|---|---|
| `BILL_AMT1` | Septembre 2005 (mois le plus récent) |
| `BILL_AMT2` | Août 2005 |
| `BILL_AMT3` | Juillet 2005 |
| `BILL_AMT4` | Juin 2005 |
| `BILL_AMT5` | Mai 2005 |
| `BILL_AMT6` | Avril 2005 |

---

### Montant des paiements effectués (avril à septembre 2005)
> Montant réellement remboursé par le client ce mois-là, en NT$.
> Toujours **≥ 0** (un remboursement ne peut pas être négatif).

| Variable | Mois correspondant |
|---|---|
| `PAY_AMT1` | Septembre 2005 |
| `PAY_AMT2` | Août 2005 |
| `PAY_AMT3` | Juillet 2005 |
| `PAY_AMT4` | Juin 2005 |
| `PAY_AMT5` | Mai 2005 |
| `PAY_AMT6` | Avril 2005 |

---

### Features dérivées (Phase 2 — Ingénierie)
> Ces variables n'existent pas dans le dataset brut.
> Elles ont été créées pour condenser l'information sur 6 mois
> et apporter un signal métier plus fort au modèle.

| Variable | Formule | Interprétation |
|---|---|---|
| `UTIL_RATIO` | `BILL_AMT1 / (LIMIT_BAL + 1)` | Taux d'utilisation du crédit — proche de 1 ou > 1 = risque élevé |
| `TOTAL_BILL` | Somme de `BILL_AMT1` à `BILL_AMT6` | Encours cumulé total sur 6 mois |
| `TOTAL_PAY` | Somme de `PAY_AMT1` à `PAY_AMT6` | Total remboursé sur 6 mois |
| `NET_DEBT` | `TOTAL_BILL − TOTAL_PAY` | Dette nette : positif = accumulation, négatif = excédent de remboursement |
| `MAX_PAY_DELAY` | `max(PAY_0, PAY_2, ..., PAY_6)` | Pire retard observé sur les 6 mois — signal le plus prédictif |
| `AVG_PAY_DELAY` | `mean(PAY_0, PAY_2, ..., PAY_6)` | Retard moyen — distingue incident isolé vs comportement chronique |
| `REPAY_RATIO` | `TOTAL_PAY / (TOTAL_BILL + 1)` | Taux de remboursement global : `1` = tout remboursé, `< 1` = dette croissante |

---

### 1.2 Validation de Schéma (équivalent Pandera/Great Expectations)

````python
# ── Règles de validation métier ──────────────────────────────────────────────
SCHEMA_RULES = {
    'AGE'      : {'min': 18, 'max': 100, 'severity': 'HIGH'},
    'SEX'      : {'allowed': [1, 2], 'severity': 'MEDIUM'},
    'EDUCATION': {'allowed': [0, 1, 2, 3, 4, 5, 6], 'severity': 'MEDIUM'},
    'MARRIAGE' : {'allowed': [0, 1, 2, 3], 'severity': 'LOW'},
    'LIMIT_BAL': {'min': 0, 'max': 5_000_000, 'severity': 'HIGH'},
    'DEFAULT'  : {'allowed': [0, 1], 'severity': 'CRITICAL'},
}
PAY_AMT_COLS = [f'PAY_AMT{i}' for i in range(1, 7)]

violations = []

for col, rules in SCHEMA_RULES.items():
    if 'min' in rules or 'max' in rules:
        lo = rules.get('min', -np.inf)
        hi = rules.get('max', np.inf)
        bad = (~df[col].between(lo, hi)).sum()
        if bad > 0:
            violations.append({'col': col, 'rule': f'[{lo}, {hi}]',
                                'n_violations': bad, 'severity': rules['severity']})
    if 'allowed' in rules:
        bad = (~df[col].isin(rules['allowed'])).sum()
        if bad > 0:
            violations.append({'col': col, 'rule': f'∈ {rules["allowed"]}',
                                'n_violations': bad, 'severity': rules['severity']})

for c in PAY_AMT_COLS:
    bad = (df[c] < 0).sum()
    if bad > 0:
        violations.append({'col': c, 'rule': '≥ 0', 'n_violations': bad, 'severity': 'HIGH'})

# ── Catégories non documentées ─────────────────────────────────────────────
edu_undoc = df['EDUCATION'].isin([0, 5, 6]).sum()
mar_undoc = df['MARRIAGE'].isin([0]).sum()
violations.append({'col': 'EDUCATION', 'rule': '0,5,6 non documenté',
                   'n_violations': edu_undoc, 'severity': 'LOW'})
violations.append({'col': 'MARRIAGE', 'rule': '0 non documenté',
                   'n_violations': mar_undoc, 'severity': 'LOW'})

vdf = pd.DataFrame(violations)
color_map = {'CRITICAL': '🔴', 'HIGH': '🟠', 'MEDIUM': '🟡', 'LOW': '🟢'}
vdf['icon'] = vdf['severity'].map(color_map)
print("=== RAPPORT DE VALIDATION DE SCHÉMA ===\n")
for _, row in vdf.iterrows():
    status = "⚠️ VIOLATION" if row['n_violations'] > 0 else "✅ OK"
    print(f"{row['icon']} [{row['severity']:8s}] {row['col']:12s} | règle: {row['rule']:25s} | {status} ({row['n_violations']} lignes)")
````

### 1.2 Validation de Schéma

Le validateur a vérifié **6 colonnes critiques** + les 6 colonnes `PAY_AMT*`
contre les règles métier définies. Voici le rapport complet :

| Sévérité | Colonne | Règle vérifiée | Résultat | Lignes affectées |
|---|---|---|---|---|
| 🟢 LOW | `EDUCATION` | Valeurs 0, 5, 6 non documentées | ⚠️ VIOLATION | **345 lignes** |
| 🟢 LOW | `MARRIAGE` | Valeur 0 non documentée | ⚠️ VIOLATION | **54 lignes** |
| 🟠 HIGH | `AGE` | Entre [18, 100] | ✅ OK | 0 |
| 🟠 HIGH | `LIMIT_BAL` | Entre [0, 5 000 000] | ✅ OK | 0 |
| 🟡 MEDIUM | `SEX` | ∈ {1, 2} | ✅ OK | 0 |
| 🔴 CRITICAL | `DEFAULT` | ∈ {0, 1} | ✅ OK | 0 |

> ✅ **Bonne nouvelle** : aucune violation critique ou haute sévérité.
> Les colonnes métier essentielles (`DEFAULT`, `AGE`, `LIMIT_BAL`, `SEX`)
> sont toutes propres et conformes au schéma attendu.

---

**Analyse des violations détectées :**

**`EDUCATION` — 345 lignes avec valeurs {0, 5, 6}**
La documentation UCI ne définit que 4 catégories officielles :
`1` = graduate school, `2` = university, `3` = high school, `4` = others.
Les codes 0, 5 et 6 sont **non documentés** mais présents dans le dataset.
→ **Décision** : ces lignes seront remappées vers la catégorie `4` (others)
pour éviter d'introduire une catégorie fantôme dans le modèle.

**`MARRIAGE` — 54 lignes avec valeur {0}**
Le schéma officiel définit `1` = married, `2` = single, `3` = others.
La valeur `0` n'est pas documentée.
→ **Décision** : remappage vers `3` (others), représente seulement **0.18%**
du dataset — impact négligeable sur les performances.

---

> 💡 **Conclusion schéma** : le dataset est globalement propre.
> Les deux violations sont de **faible sévérité** (LOW) et concernent
> des codes de catégories ambiguës, non des erreurs de saisie.
> Le pipeline peut continuer sans blocage.

### 1.3 Analyse de Distribution & Outliers (Z-score + IQR)

````python
numeric_cols = ['LIMIT_BAL', 'AGE'] + [f'BILL_AMT{i}' for i in range(1,7)] + [f'PAY_AMT{i}' for i in range(1,7)]

outlier_report = []
for col in numeric_cols:
    series = df[col].dropna()
    z_out = int((np.abs(stats.zscore(series)) > 3).sum())
    Q1, Q3 = series.quantile(0.25), series.quantile(0.75)
    IQR = Q3 - Q1
    iqr_out = int(((series < Q1-1.5*IQR) | (series > Q3+1.5*IQR)).sum())
    outlier_report.append({'feature': col, 'Z-score (>3σ)': z_out,
                           'IQR (1.5×)': iqr_out, 'pct_iqr': round(iqr_out/len(df)*100, 1)})

out_df = pd.DataFrame(outlier_report).set_index('feature')
print("=== RAPPORT OUTLIERS ===")
print(out_df.to_string())

# ── Visualisation distributions ────────────────────────────────────────────
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle("Phase 1 – Distributions clés & Outliers", fontsize=14, fontweight='bold')
key_cols = ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'PAY_AMT1', 'UTIL_RATIO' if 'UTIL_RATIO' in df.columns else 'BILL_AMT2']

for ax, col in zip(axes.flat, numeric_cols[:6]):
    Q1, Q3 = df[col].quantile(0.25), df[col].quantile(0.75)
    IQR = Q3 - Q1
    lo, hi = Q1 - 1.5*IQR, Q3 + 1.5*IQR
    ax.hist(df[col], bins=50, color='#3498db', alpha=0.7, edgecolor='white')
    ax.axvline(lo, color='#e74c3c', linestyle='--', alpha=0.8, label='IQR fence')
    ax.axvline(hi, color='#e74c3c', linestyle='--', alpha=0.8)
    ax.set_title(col, fontsize=10)
    ax.legend(fontsize=7)

axes[1][2].axis('off')
plt.tight_layout()
# plt.savefig('phase1_distributions.png', dpi=120, bbox_inches='tight')
plt.show()
print("✅ Figure sauvegardée : phase1_distributions.png")
````

> Dans un contexte de crédit, les comportements extrêmes (très gros remboursement,

> Supprimer ces lignes reviendrait à apprendre sur la population "normale"
> et à échouer précisément sur les cas les plus risqués.

## ⚗️ Phase 2 : Validation de l'Ingénierie des Caractéristiques

Chaque transformation est testée et vérifiée avant d'entrer dans le modèle.

### 2.1 Nettoyage & Encodage des Catégories Non Documentées

````python
# Harmonisation des catégories non documentées
df['EDUCATION'] = df['EDUCATION'].replace({0: 4, 5: 4, 6: 4})  # → 'other'
df['MARRIAGE']  = df['MARRIAGE'].replace({0: 3})                 # → 'other'
print("✅ Catégories remappées : EDUCATION {0,5,6}→4  |  MARRIAGE {0}→3")
````

### 2.2 Création des Features Métier

````python
# ── Features dérivées ──────────────────────────────────────────────────────
df['UTIL_RATIO']    = df['BILL_AMT1'] / (df['LIMIT_BAL'] + 1)   # Taux utilisation crédit
# Le +1 évite la division par zéro. Mesure à quel point le client utilise son crédit disponible
df['TOTAL_BILL']    = df[[f'BILL_AMT{i}' for i in range(1,7)]].sum(axis=1)
# Encours total sur 6 mois — vision globale de la dette accumulée
df['TOTAL_PAY']     = df[[f'PAY_AMT{i}' for i in range(1,7)]].sum(axis=1)
# Total remboursé sur 6 mois
df['NET_DEBT']      = df['TOTAL_BILL'] - df['TOTAL_PAY']          # Dette nette
# Dette nette réelle : positif = le client s'endette, négatif = il rembourse plus qu'il ne doit
df['MAX_PAY_DELAY'] = df[['PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6']].max(axis=1)
# Pire retard observé sur les 6 mois — signal fort de risque
df['AVG_PAY_DELAY'] = df[['PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6']].mean(axis=1)
# Retard moyen — distingue un incident isolé d'un comportement chronique
df['REPAY_RATIO']   = df['TOTAL_PAY'] / (df['TOTAL_BILL'] + 1)   # Taux remboursement
# Ratio de remboursement : 1.0 = rembourse tout, < 1 = s'endette progressivement

# ── CORRECTION : Clipping de REPAY_RATIO ──────────────────────────────────
# TOTAL_BILL peut être négatif (crédit excédentaire) → ratio explose ou devient négatif
# Un ratio > 10 n'a pas de sens métier → on borne à [0, 10]
df['REPAY_RATIO'] = df['REPAY_RATIO'].clip(lower=0, upper=10)
print(f"✅ REPAY_RATIO après clipping : range [{df['REPAY_RATIO'].min():.2f}, {df['REPAY_RATIO'].max():.2f}]")

new_features = ['UTIL_RATIO', 'TOTAL_BILL', 'TOTAL_PAY', 'NET_DEBT',
                'MAX_PAY_DELAY', 'AVG_PAY_DELAY', 'REPAY_RATIO']
print(f"✅ {len(new_features)} nouvelles features créées :")
for f in new_features:
    print(f"   • {f:20s}: range [{df[f].min():.1f}, {df[f].max():.1f}], nulls={df[f].isnull().sum()}")
````

### 2.2 Création des Features Métier

**7 nouvelles features dérivées** ont été créées pour enrichir le signal prédictif.
Toutes sont complètes : `nulls = 0` sur les 30 000 observations.

---

**Résultats et interprétation par feature :**

#### `UTIL_RATIO` — Taux d'utilisation du crédit
- **Plage** : [-0.6 ; 6.5]
- Les valeurs **négatives** (crédit excédentaire sur BILL_AMT1) sont possibles et légitimes.
- Les valeurs **> 1.0** signifient que le client a dépassé sa limite de crédit
  (pénalités probables). Une valeur de 6.5 indique un dépassement massif.
- C'est l'un des indicateurs les plus prédictifs du risque de défaut :
  un ratio proche de 1 ou supérieur est un signal d'alerte fort.

#### `TOTAL_BILL` — Encours total sur 6 mois
- **Plage** : [-336 259 ; 5 263 883 NT$]
  
- La valeur négative minimale confirme que certains clients ont un crédit
  excédentaire sur tous leurs relevés.
- Un encours cumulé de 5.2M NT$ pour un seul client indique des cas extrêmes
  qui seront gérés par la normalisation StandardScaler.

#### `TOTAL_PAY` — Total remboursé sur 6 mois
- **Plage** : [0 ; 3 764 066 NT$]
- Minimum à **0** : certains clients n'ont effectué aucun remboursement
  sur 6 mois — signal critique de défaut imminent.
- La distribution est très asymétrique : la majorité rembourse peu,
  quelques clients soldent d'importantes dettes.

#### `NET_DEBT` — Dette nette
- **Plage** : [-2 671 514 ; 4 116 080 NT$]

  
- **Négatif** → le client rembourse plus qu'il ne consomme : profil sain.
- **Positif** → la dette s'accumule mois après mois : profil à risque.
- Un `NET_DEBT` de 4.1M NT$ représente une accumulation de dette
  sur 6 mois sans remboursement significatif.

#### `MAX_PAY_DELAY` — Pire retard de paiement
- **Plage** : [-2 ; 8 mois]
- `-2` = paiement en avance ou aucun débit, `0` = paiement ponctuel,
  `8` = 8 mois de retard consécutif — défaut quasi-certain.
- C'est typiquement la **feature la plus importante** dans les modèles
  de scoring crédit, confirmé plus tard par l'importance RF.

#### `AVG_PAY_DELAY` — Retard moyen
- **Plage** : [-2 ; 6 mois]
- Complète `MAX_PAY_DELAY` : un retard max de 8 sur un seul mois
  est moins grave qu'une moyenne de 6 sur tous les mois.
- Permet de distinguer l'**incident ponctuel** du **comportement chronique**.

#### `REPAY_RATIO` — Taux de remboursement
- **Plage brute** : [-589 ; 162 000]
- La plage extrême s'explique par le `+1` au dénominateur : quand
  `TOTAL_BILL ≈ 0` et `TOTAL_PAY > 0`, le ratio explose. De même,
  quand `TOTAL_BILL` est négatif (crédit excédentaire), le ratio
  devient négatif — ce qui n'a pas de sens métier.

> ✅ **Correction appliquée** : clipping dans l'intervalle `[0, 10]`
> ```python
> df['REPAY_RATIO'] = df['REPAY_RATIO'].clip(lower=0, upper=10)
> ```
> **Après clipping** : range [0.00 ; 10.00], nulls = 0
>
> Cette borne est justifiée métier : un client qui rembourse plus de
> 10× son encours est un cas extrême sans valeur prédictive
> supplémentaire. En pratique, les valeurs utiles se concentrent
> entre **0 et 2** — `0` signifiant aucun remboursement (risque
> maximal) et `1` signifiant remboursement intégral (risque minimal).

---

> 💡 **Pourquoi ces features ?**
> Les colonnes brutes `BILL_AMT*` et `PAY_AMT*` donnent des snapshots mensuels
> isolés. Ces 7 features agrègent le **comportement sur 6 mois** et créent
> des ratios que le modèle ne peut pas construire implicitement depuis les
> colonnes séparées, surtout pour la Régression Logistique qui n'apprend
> pas d'interactions automatiquement.

### 2.3 Tests Unitaires du Pipeline de Transformation

````python
# ── Tests unitaires ────────────────────────────────────────────────────────
tests = []

def run_test(name, condition, detail=""):
    status = "✅ PASS" if condition else "❌ FAIL"
    tests.append({'test': name, 'status': status, 'detail': detail})
    print(f"{status}  {name}  {detail}")

print("=== TESTS UNITAIRES DU PIPELINE ===\n")

# Aucun null dans les features dérivées
for feat in ['UTIL_RATIO','TOTAL_BILL','TOTAL_PAY','NET_DEBT','MAX_PAY_DELAY','REPAY_RATIO']:
    run_test(f"No nulls in {feat}", df[feat].isnull().sum() == 0)

# TOTAL_PAY ne peut être négatif (somme de PAY_AMT*)
run_test("TOTAL_PAY >= 0", (df['TOTAL_PAY'] >= 0).all())

# MAX_PAY_DELAY dans plage connue
run_test("MAX_PAY_DELAY ∈ [-2, 9]", df['MAX_PAY_DELAY'].between(-2, 9).all())

# UTIL_RATIO = BILL_AMT1 / (LIMIT_BAL + 1) sur ligne test
test_row = pd.Series({'BILL_AMT1': 10000, 'LIMIT_BAL': 50000})
expected = 10000 / 50001
actual = test_row['BILL_AMT1'] / (test_row['LIMIT_BAL'] + 1)
run_test("UTIL_RATIO formula exact", abs(actual - expected) < 1e-6, f"(expected {expected:.6f})")

# REPAY_RATIO entre 0 et borne haute raisonnable
run_test("REPAY_RATIO >= 0", (df['REPAY_RATIO'] >= 0).all())

# ── Vérification fuite de données ─────────────────────────────────────────
FEATURE_COLS = [c for c in df.columns if c not in ['ID', 'DEFAULT']]
run_test("Fuite: DEFAULT absent des features", 'DEFAULT' not in FEATURE_COLS)
run_test("Fuite: ID absent des features", 'ID' not in FEATURE_COLS)
run_test("Fuite: date_of_default absente", 'date_of_default' not in FEATURE_COLS)

n_pass = sum(1 for t in tests if '✅' in t['status'])
print(f"\n{'─'*45}")
print(f"Résultat : {n_pass}/{len(tests)} tests passés")
````

### 2.3 Tests Unitaires du Pipeline de Transformation

**13 tests automatisés** ont été exécutés pour valider l'intégrité

> ✅ Le clipping appliqué sur `REPAY_RATIO` à l'étape précédente

### Tests Additionnels — Contrôle Qualité Approfondi

## 🤖 Phase 3 : Validation du Modèle

Trois modèles comparés via validation croisée stratifiée, avec analyse des métriques métier et de l'équité.

### 3.1 Split Train / Validation / Test

````python
FEATURE_COLS = [c for c in df.columns if c not in ['ID', 'DEFAULT']]
X = df[FEATURE_COLS]
y = df['DEFAULT']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=42)
X_train, X_val, y_train, y_val   = train_test_split(X_train, y_train, test_size=0.125, stratify=y_train, random_state=42)

print(f"Train : {X_train.shape[0]:,} lignes ({y_train.mean()*100:.1f}% défaut)")
print(f"Val   : {X_val.shape[0]:,} lignes ({y_val.mean()*100:.1f}% défaut)")
print(f"Test  : {X_test.shape[0]:,} lignes ({y_test.mean()*100:.1f}% défaut)")
````

### 3.1 Split Train / Validation / Test

#### Stratégie de découpage

Un **double split stratifié** a été appliqué pour créer 3 jeux
totalement indépendants :

| Jeu | Lignes | % du total | Taux défaut | Rôle |
|---|---|---|---|---|
| **Train** | 21 000 | 70% | 22.1% | Apprentissage du modèle |
| **Validation** | 3 000 | 10% | 22.1% | Ajustement des hyperparamètres |
| **Test** | 6 000 | 20% | 22.1% | Évaluation finale non biaisée |
| **Total** | 30 000 | 100% | 22.1% | — |

---

#### Validation de la stratification

Le taux de défaut est **identique dans les 3 splits : 22.1%**
C'est le résultat direct du paramètre `stratify=y` dans
`train_test_split()`. Sans ce paramètre, un split aléatoire
pourrait produire par exemple 19% de défauts dans le test
et 23% dans le train — le modèle apprendrait sur une
distribution différente de celle qu'il évaluerait.

> ✅ La stratification garantit que le déséquilibre de classes
> (3.5:1) est **représenté de façon identique** dans chaque
> jeu, rendant les métriques comparables entre splits.

---

#### Rôle de chaque jeu

**Train (21 000 lignes)**
Le modèle ajuste ses paramètres internes uniquement sur ce jeu.
21 000 observations est suffisant pour apprendre des patterns
robustes avec les 3 algorithmes utilisés.

**Validation (3 000 lignes)**
Jeu intermédiaire utilisé pendant le développement pour
comparer les modèles et détecter le surapprentissage
**sans contaminer le jeu de test**. Un écart AUC
train/val > 5 pts sera considéré comme un signal de
surapprentissage.

**Test (6 000 lignes)**
Jeu verrouillé — il ne doit être utilisé qu'**une seule fois**
pour l'évaluation finale. Toute décision de modélisation
prise en observant le jeu de test introduirait un biais
d'optimisme dans les métriques rapportées.

### Vérifications Post-Split

````python
# ── Amélioration 1 : vérification explicite de la stratification ──────────
for nom, y_split in [('Train', y_train), ('Val', y_val), ('Test', y_test)]:
    taux = y_split.mean() * 100
    assert abs(taux - 22.1) < 0.5, f"Stratification échouée sur {nom} : {taux:.1f}%"
    print(f"✅ {nom:5s} : {taux:.1f}% défaut — stratification OK")

# ── Amélioration 2 : vérification d'absence de fuite entre les splits ─────
# S'assure qu'aucun index client n'apparaît dans 2 splits différents
train_idx = set(X_train.index)
val_idx   = set(X_val.index)
test_idx  = set(X_test.index)

assert len(train_idx & val_idx)  == 0, "⚠️ Fuite train/val détectée !"
assert len(train_idx & test_idx) == 0, "⚠️ Fuite train/test détectée !"
assert len(val_idx   & test_idx) == 0, "⚠️ Fuite val/test détectée !"
print("✅ Aucune fuite entre les 3 splits — index totalement disjoints")
````

### Vérifications Post-Split

#### Validation de la Stratification
Les 3 assertions confirment que le taux de défaut est **identique
dans chaque split à 22.1%** — aucune déviation détectée.

| Split | Taux défaut attendu | Taux défaut obtenu | Statut |
|---|---|---|---|
| Train | 22.1% | 22.1% | ✅ OK |
| Val | 22.1% | 22.1% | ✅ OK |
| Test | 22.1% | 22.1% | ✅ OK |

> ✅ La stratification est **parfaite** — le déséquilibre de classes
> 3.5:1 est rigoureusement préservé dans les 3 jeux. Le modèle
> apprend et est évalué sur des distributions **strictement identiques**,
> ce qui rend les comparaisons de métriques entre splits totalement fiables.

---

#### Validation de la Disjonction des Index
```
Train ∩ Val  = 0 client commun  ✅
Train ∩ Test = 0 client commun  ✅
Val   ∩ Test = 0 client commun  ✅
```

Les 3 jeux sont **totalement disjoints** — aucun client n'apparaît
dans plus d'un split. C'est une garantie fondamentale contre
la **contamination des données** :

| Risque évité | Conséquence si non vérifié |
|---|---|
| Client en train **et** en test | Le modèle aurait déjà "vu" certains cas de test → métriques artificiellement gonflées |
| Client en train **et** en val | La validation croisée perdrait son objectivité → sélection de modèle biaisée |
| Client en val **et** en test | L'évaluation finale ne serait plus indépendante |

> ✅ **Conclusion** : le pipeline de split est **doublement validé** —
> stratification correcte ET disjonction garantie.
> Les performances mesurées sur le jeu de test reflètent
> fidèlement ce qu'obtiendra le modèle en production
> sur de nouveaux clients jamais vus.

### 3.2 Entraînement & Cross-Validation Stratifiée (5 folds)

````python
models = {
    'Logistic\nRegression': Pipeline([
        ('scaler', StandardScaler()),
        ('clf', LogisticRegression(class_weight='balanced', max_iter=500, C=0.1))
    ]),
    'Random\nForest': Pipeline([
        ('scaler', StandardScaler()),
        ('clf', RandomForestClassifier(n_estimators=100, class_weight='balanced',
                                       random_state=42, n_jobs=-1))
    ]),
    'Gradient\nBoosting': Pipeline([
        ('scaler', StandardScaler()),
        ('clf', GradientBoostingClassifier(n_estimators=100, learning_rate=0.05,
                                           max_depth=4, random_state=42))
    ])
}

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_results, model_metrics, trained_models = {}, {}, {}

for name, model in models.items():
    clean = name.replace('\n', ' ')
    print(f"⏳ Entraînement {clean}...")
    cv = cross_validate(model, X_train, y_train, cv=skf,
                        scoring=['roc_auc','f1','precision','recall'], n_jobs=-1)
    cv_results[name] = {k: (np.mean(v), np.std(v)) for k,v in cv.items() if k.startswith('test_')}

    model.fit(X_train, y_train)
    trained_models[name] = model

    metrics = {}
    for split, Xs, ys in [('train',X_train,y_train),('val',X_val,y_val),('test',X_test,y_test)]:
        yp = model.predict(Xs)
        yprob = model.predict_proba(Xs)[:,1]
        tn, fp, fn, tp = confusion_matrix(ys, yp).ravel()
        metrics[split] = {
            'accuracy': round(float((yp==ys).mean()), 4),
            'precision': round(float(precision_score(ys, yp)), 4),
            'recall': round(float(recall_score(ys, yp)), 4),
            'f1': round(float(f1_score(ys, yp)), 4),
            'roc_auc': round(float(roc_auc_score(ys, yprob)), 4),
            'tn':int(tn),'fp':int(fp),'fn':int(fn),'tp':int(tp),
            'business_cost': int(fn*5 + fp*1)
        }
    model_metrics[name] = metrics
    auc = metrics['test']['roc_auc']
    cv_auc = cv_results[name]['test_roc_auc']
    print(f"   CV AUC: {cv_auc[0]:.4f} ±{cv_auc[1]:.4f}  |  Test AUC: {auc}")

print("\n✅ Entraînement terminé")
````

### 3.3 Tableau de Bord des Métriques (Train / Val / Test)

````python
# ── Tableau comparatif ────────────────────────────────────────────────────
rows = []
for name, mets in model_metrics.items():
    for split, m in mets.items():
        rows.append({'Modèle': name.replace('\n',' '), 'Split': split,
                     'AUC': m['roc_auc'], 'F1': m['f1'],
                     'Recall': m['recall'], 'Precision': m['precision'],
                     'Coût Métier': m['business_cost']})

metrics_df = pd.DataFrame(rows)
print(metrics_df.pivot(index='Modèle', columns='Split',
                        values=['AUC','F1','Recall','Coût Métier']).round(4).to_string())
````

### 3.4 Matrices de Confusion & Courbes ROC

````python
fig = plt.figure(figsize=(18, 12))
gs = gridspec.GridSpec(2, 3, figure=fig)
fig.suptitle("Phase 3 – Analyse des Modèles", fontsize=15, fontweight='bold')

colors = ['#3498db', '#2ecc71', '#e74c3c']
for idx, (name, model) in enumerate(trained_models.items()):
    clean = name.replace('\n', ' ')
    yp = model.predict(X_test)
    yprob = model.predict_proba(X_test)[:,1]

    # Confusion matrix
    ax_cm = fig.add_subplot(gs[0, idx])
    cm = confusion_matrix(y_test, yp)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax_cm,
                xticklabels=['Prédit 0','Prédit 1'],
                yticklabels=['Réel 0','Réel 1'])
    ax_cm.set_title(f'{clean}\nCM (Test)', fontsize=10)

    # ROC Curve
    ax_roc = fig.add_subplot(gs[1, idx])
    for split, Xs, ys, ls in [('Train',X_train,y_train,'--'),
                               ('Val',X_val,y_val,':'),
                               ('Test',X_test,y_test,'-')]:
        fpr, tpr, _ = roc_curve(ys, model.predict_proba(Xs)[:,1])
        auc = roc_auc_score(ys, model.predict_proba(Xs)[:,1])
        ax_roc.plot(fpr, tpr, linestyle=ls, label=f'{split} AUC={auc:.3f}')
    ax_roc.plot([0,1],[0,1],'k--', alpha=0.3)
    ax_roc.set_xlabel('FPR'); ax_roc.set_ylabel('TPR')
    ax_roc.set_title(f'{clean}\nROC Curves', fontsize=10)
    ax_roc.legend(fontsize=8)

plt.tight_layout()
plt.savefig('phase3_model_analysisvf.png', dpi=120, bbox_inches='tight')
plt.show()
````

### 3.5 Analyse d'Équité (Fairness)

````python
best_name = max(model_metrics, key=lambda n: model_metrics[n]['test']['roc_auc'])
best_model = trained_models[best_name]
print(f"Meilleur modèle : {best_name.replace(chr(10),' ')}")

test_df = X_test.copy()
test_df['DEFAULT'] = y_test.values
test_df['PRED'] = best_model.predict(X_test)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Phase 3 – Analyse d'Équité (Fairness)", fontsize=13, fontweight='bold')

for ax, (grp_col, title, labels) in zip(axes, [
    ('SEX',     'Équité par Genre',       {1:'Homme',2:'Femme'}),
    ('AGE_GRP', 'Équité par Tranche d\'Âge', None)
]):
    if grp_col == 'AGE_GRP':
        test_df['AGE_GRP'] = pd.cut(test_df['AGE'], bins=[0,30,45,100],
                                     labels=['Jeune','Adulte','Senior'])
    grps, recalls, fprs, fnrs = [], [], [], []
    for grp in sorted(test_df[grp_col].dropna().unique()):
        sub = test_df[test_df[grp_col]==grp]
        lbl = labels.get(grp, str(grp)) if labels else str(grp)
        grps.append(lbl)
        recalls.append(recall_score(sub['DEFAULT'], sub['PRED'], zero_division=0))
        fprs.append((sub[(sub['DEFAULT']==0)&(sub['PRED']==1)].shape[0]) /
                    max(sub[sub['DEFAULT']==0].shape[0],1))
        fnrs.append((sub[(sub['DEFAULT']==1)&(sub['PRED']==0)].shape[0]) /
                    max(sub[sub['DEFAULT']==1].shape[0],1))

    x = np.arange(len(grps))
    w = 0.25
    ax.bar(x-w, recalls, w, label='Recall', color='#3498db')
    ax.bar(x,   fprs,    w, label='FPR',    color='#f39c12')
    ax.bar(x+w, fnrs,    w, label='FNR',    color='#e74c3c')
    ax.set_xticks(x); ax.set_xticklabels(grps)
    ax.set_ylim(0, 1); ax.set_ylabel('Taux'); ax.set_title(title)
    ax.legend()
    ax.axhline(0.05, color='gray', linestyle=':', alpha=0.5, label='Seuil 5%')

plt.tight_layout()
plt.savefig('phase3_fairnessvf.png', dpi=120, bbox_inches='tight')
plt.show()

# ── Tableau équité ─────────────────────────────────────────────────────────
for grp_col, labels in [('SEX',{1:'Homme',2:'Femme'}), ('AGE_GRP', None)]:
    print(f"\n--- Équité par {grp_col} ---")
    for grp in sorted(test_df[grp_col].dropna().unique()):
        sub = test_df[test_df[grp_col]==grp]
        lbl = labels.get(grp, str(grp)) if labels else str(grp)
        r = recall_score(sub['DEFAULT'], sub['PRED'], zero_division=0)
        p = precision_score(sub['DEFAULT'], sub['PRED'], zero_division=0)
        fnr = (sub[(sub['DEFAULT']==1)&(sub['PRED']==0)].shape[0]) / max(sub[sub['DEFAULT']==1].shape[0],1)
        print(f"  {lbl:8s} | n={len(sub):5,} | recall={r:.3f} | precision={p:.3f} | FNR={fnr:.3f}")
````

## 📡 Phase 4 : Monitoring & Drift Simulation

Un modèle performant aujourd'hui peut être obsolète demain. Cette phase simule un changement de distribution et implémente un détecteur statistique.

### 4.1 Simulation de Data Drift (choc économique)

````python
# ── Simulation : choc d'inflation + resserrement crédit ────────────────────
np.random.seed(42)
X_ref   = X_test.copy()
X_drift = X_test.copy()

# Choc économique simulé
X_drift['LIMIT_BAL'] = X_drift['LIMIT_BAL'] * 0.70   # resserrement crédit -30%
X_drift['BILL_AMT1'] = X_drift['BILL_AMT1'] * 1.40   # inflation +40%
X_drift['PAY_AMT1']  = X_drift['PAY_AMT1']  * 0.80   # stress de remboursement -20%
X_drift['AGE']       = np.clip(X_drift['AGE'] + np.random.normal(3, 1, len(X_drift)), 18, 90)
X_drift['UTIL_RATIO']= X_drift['BILL_AMT1'] / (X_drift['LIMIT_BAL'] + 1)

print("=== SIMULATION DE CHOC ÉCONOMIQUE ===")
print("Scénario : inflation +40%, resserrement crédit -30%, stress remboursements -20%\n")

drift_cols = ['LIMIT_BAL','BILL_AMT1','PAY_AMT1','AGE','UTIL_RATIO']
drift_report = []
for col in drift_cols:
    ks_stat, ks_pval = stats.ks_2samp(X_ref[col], X_drift[col])
    detected = ks_pval < 0.05
    pct = (X_drift[col].mean() - X_ref[col].mean()) / abs(X_ref[col].mean()) * 100
    drift_report.append({
        'Feature': col, 'KS Stat': round(ks_stat,4),
        'p-value': round(ks_pval,6), 'Drift Détecté': '🚨 OUI' if detected else '✅ NON',
        'Δ Moyenne': f"{pct:+.1f}%"
    })
    print(f"{'🚨' if detected else '✅'} {col:15s} KS={ks_stat:.4f}  p={ks_pval:.2e}  Δmoy={pct:+.1f}%")

drift_df = pd.DataFrame(drift_report)
````

### 4.2 Visualisation du Drift

````python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle("Phase 4 – Data Drift (KS Test)", fontsize=14, fontweight='bold')

for ax, col in zip(axes.flat, drift_cols):
    ax.hist(X_ref[col],   bins=50, alpha=0.6, color='#3498db', label='Référence', density=True)
    ax.hist(X_drift[col], bins=50, alpha=0.6, color='#e74c3c', label='Drift',     density=True)
    ks, pv = stats.ks_2samp(X_ref[col], X_drift[col])
    status = "🚨 DRIFT" if pv < 0.05 else "✅ STABLE"
    ax.set_title(f"{col}\nKS={ks:.3f} p={pv:.2e} {status}", fontsize=9)
    ax.legend(fontsize=7)

axes[1][2].axis('off')
plt.tight_layout()
plt.savefig('phase4_drift.pngvf', dpi=120, bbox_inches='tight')
plt.show()

# ── Dégradation de performance post-drift ─────────────────────────────────
print("\n=== DÉGRADATION DES PERFORMANCES APRÈS DRIFT ===")
drift_features = [c for c in X_drift.columns if c in FEATURE_COLS]
X_d_sub = X_drift[drift_features] if set(drift_features) == set(FEATURE_COLS) else X_drift[FEATURE_COLS]

# Reconstruct drift features
X_drift_full = X_drift.copy()
for c in FEATURE_COLS:
    if c not in X_drift_full.columns:
        X_drift_full[c] = X_test[c]
X_drift_full = X_drift_full[FEATURE_COLS]

auc_ref   = roc_auc_score(y_test, best_model.predict_proba(X_test[FEATURE_COLS])[:,1])
try:
    auc_drift = roc_auc_score(y_test, best_model.predict_proba(X_drift_full)[:,1])
    print(f"AUC Référence  : {auc_ref:.4f}")
    print(f"AUC Post-Drift : {auc_drift:.4f}")
    print(f"Dégradation    : {(auc_ref - auc_drift)*100:.2f} pts")
    if auc_ref - auc_drift > 0.02:
        print("🚨 ALERTE : dégradation >2 pts → réentraînement recommandé")
except Exception as e:
    print(f"Note: {e}")
````
