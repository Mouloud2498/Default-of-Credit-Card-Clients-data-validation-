## Description du jeu de données

Le jeu de données **Default of Credit Card Clients** porte sur des clients détenteurs de cartes de crédit à Taïwan et a pour objectif principal d’étudier le **risque de défaut de paiement**. Il est largement utilisé dans les projets de **classification binaire**, de **scoring de crédit** et d’**analyse du risque financier**, car il permet d’examiner les facteurs associés à la probabilité qu’un client ne rembourse pas ses obligations à temps.

Ce dataset s’inscrit dans un contexte économique réel marqué par une forte crise de l’endettement liée aux cartes de crédit. À cette période, l’octroi massif de crédit à des clients parfois peu solvables, combiné à une surconsommation et à l’accumulation de dettes, a mis en évidence la nécessité pour les institutions financières et les autorités de mieux identifier les profils à risque. Dans ce cadre, l’analyse des historiques de paiement et des caractéristiques socio-démographiques permet de construire des systèmes capables d’anticiper les comportements de défaut.

Le fichier contient **30 000 observations**, chacune représentant un client, et **25 colonnes** dans le fichier CSV réel. Parmi ces colonnes, on retrouve :
- une colonne d’identification (`ID`) ;
- **23 variables explicatives** décrivant le profil financier, démographique et comportemental du client ;
- **1 variable cible** indiquant si le client fera défaut de paiement le mois suivant.

La variable cible est :
- **`default.payment.next.month`**
  - `0` : le client ne fait pas défaut ;
  - `1` : le client fait défaut.

Il s’agit donc d’un problème de **classification supervisée binaire**, où l’objectif est de prédire, à partir des informations passées d’un client, s’il présente ou non un risque de défaut futur.

### Structure des variables

Les variables du dataset peuvent être regroupées en plusieurs grandes catégories :

#### 1. Informations financières générales
- **`LIMIT_BAL`** : montant total du crédit accordé au client, exprimé en dollars taïwanais (NT$).  
  Cette variable représente la limite de crédit disponible pour le client et reflète indirectement son profil financier et sa capacité de crédit.

#### 2. Variables démographiques
- **`SEX`** : sexe du client ;
- **`EDUCATION`** : niveau d’éducation ;
- **`MARRIAGE`** : statut matrimonial ;
- **`AGE`** : âge du client en années.

Ces variables permettent d’explorer si certains profils socio-démographiques présentent un risque de défaut plus élevé que d’autres.

#### 3. Historique des statuts de remboursement
- **`PAY_0`, `PAY_2`, `PAY_3`, `PAY_4`, `PAY_5`, `PAY_6`**

Ces variables décrivent le **statut de remboursement mensuel** du client pour les mois précédents. Elles indiquent si le client a payé à temps ou s’il a accumulé un certain nombre de mois de retard. Elles constituent un groupe de variables très important, car elles traduisent directement le comportement récent du client face à ses obligations de paiement.

#### 4. Montants de facturation mensuels
- **`BILL_AMT1` à `BILL_AMT6`**

Ces colonnes représentent les **montants facturés** sur les relevés mensuels de la carte de crédit pour plusieurs mois consécutifs. Elles permettent de mesurer le niveau d’endettement du client dans le temps.

#### 5. Montants remboursés mensuellement
- **`PAY_AMT1` à `PAY_AMT6`**

Ces variables indiquent les **montants effectivement remboursés** par le client sur les mêmes périodes. Elles complètent l’historique de paiement en montrant l’effort réel de remboursement fourni par le client.

### Intérêt analytique du dataset

Ce jeu de données est particulièrement riche, car il combine :
- des variables **démographiques** ;
- des variables **financières** ;
- des variables **temporelles** liées au comportement de paiement.

Cette diversité permet d’étudier plusieurs dimensions du risque de défaut :
- le poids du profil personnel du client ;
- l’impact du niveau de crédit accordé ;
- le rôle des montants facturés et remboursés ;
- surtout, l’importance des **retards de paiement passés** comme indicateurs avancés du défaut futur.

En pratique, ce dataset est pertinent pour :
- construire un **modèle de score de crédit** ;
- identifier les **variables les plus prédictives** du défaut ;
- détecter les clients à haut risque ;
- proposer des stratégies de prévention ou de gestion du risque.

L’objectif est maintenant double :

## Validation de schéma

La validation de schéma consiste à définir un ensemble de règles permettant de vérifier que les colonnes du jeu de données respectent une structure cohérente, des types corrects et des domaines de valeurs plausibles.

Dans ce projet, les règles de validation ne sont pas définies à partir d’une documentation externe, mais à partir du **contenu réel du fichier CSV analysé**. L’objectif est donc de contrôler que les données restent cohérentes avec ce qui est effectivement observé dans le dataset.

Les principales règles retenues sont les suivantes :

- `ID` doit être **unique** et **strictement positif** ;
- `LIMIT_BAL` doit être **strictement positif** ;
- `SEX` doit appartenir aux **valeurs observées dans le fichier**, soit : `1` et `2` ;
- `EDUCATION` doit appartenir aux **valeurs observées**, soit : `0, 1, 2, 3, 4, 5, 6` ;
- `MARRIAGE` doit appartenir aux **valeurs observées**, soit : `0, 1, 2, 3` ;
- `AGE` doit rester dans une plage réaliste, ici comprise entre **18 et 100 ans** ;
- les variables de statut de paiement `PAY_*` doivent rester dans l’intervalle observé dans le dataset, soit de **-2 à 8** ;
- les variables de remboursement `PAY_AMT1` à `PAY_AMT6` ne doivent pas être négatives ;
- la variable cible `default.payment.next.month` doit rester **binaire**, avec les valeurs `0` et `1`.

Concernant les colonnes `BILL_AMT1` à `BILL_AMT6`, aucune contrainte de non-négativité n’est imposée dans cette phase. En effet, des valeurs négatives sont réellement présentes dans le fichier et peuvent correspondre à des situations telles que des ajustements comptables, des remboursements, des trop-payés ou des soldes créditeurs.

Cette validation permet donc de s’assurer que les données sont techniquement cohérentes avec la structure réellement observée dans le jeu de données avant de poursuivre l’analyse.

```python
observed_domains = {
    "SEX": sorted(df["SEX"].unique().tolist()),
    "AGE": sorted(df["AGE"].unique().tolist()),
    "EDUCATION": sorted(df["EDUCATION"].unique().tolist()),
    "MARRIAGE": sorted(df["MARRIAGE"].unique().tolist()),
    "PAY_0": sorted(df["PAY_0"].unique().tolist()),
    "PAY_2": sorted(df["PAY_2"].unique().tolist()),
    "PAY_3": sorted(df["PAY_3"].unique().tolist()),
    "PAY_4": sorted(df["PAY_4"].unique().tolist()),
    "PAY_5": sorted(df["PAY_5"].unique().tolist()),
    "PAY_6": sorted(df["PAY_6"].unique().tolist()),
    "TARGET": sorted(df[target].unique().tolist())
}

for key, value in observed_domains.items():
    print(f"{key} : {value}")
```

Les domaines affichés ci-dessus correspondent aux valeurs réellement présentes dans le dataset. Ils servent de base pour définir des règles de validation adaptées au fichier étudié.

Cette approche permet de construire un contrôle qualité plus robuste dans le contexte d’un projet exploratoire ou pédagogique, en évitant de rejeter des valeurs qui existent réellement dans les données.

```python
technical_schema_columns = {
    "ID": Column(int, checks=Check.ge(1), nullable=False, unique=True, coerce=True),
    "LIMIT_BAL": Column(float, checks=Check.gt(0), nullable=False, coerce=True),
    "SEX": Column(int, checks=Check.isin(sorted(df["SEX"].unique().tolist())), nullable=False, coerce=True),
    "EDUCATION": Column(int, checks=Check.isin(sorted(df["EDUCATION"].unique().tolist())), nullable=False, coerce=True),
    "MARRIAGE": Column(int, checks=Check.isin(sorted(df["MARRIAGE"].unique().tolist())), nullable=False, coerce=True),
    "AGE": Column(int, checks=Check.in_range(18, 100), nullable=False, coerce=True),
    target: Column(int, checks=Check.isin([0, 1]), nullable=False, coerce=True)
}

for col in pay_status_cols:
    technical_schema_columns[col] = Column(
        int,
        checks=Check.in_range(df[col].min(), df[col].max()),
        nullable=False,
        coerce=True
    )

for col in bill_cols:
    technical_schema_columns[col] = Column(
        float,
        nullable=False,
        coerce=True
    )

for col in pay_amt_cols:
    technical_schema_columns[col] = Column(
        float,
        checks=Check.ge(0),
        nullable=False,
        coerce=True
    )

technical_schema = DataFrameSchema(technical_schema_columns, strict=True)

try:
    technical_schema.validate(df, lazy=True)
    print("Validation technique réussie : aucune violation détectée.")
except pa.errors.SchemaErrors as e:
    print("Violations détectées dans la validation technique :")
    display(e.failure_cases)
```

## Interprétation de la validation de schéma

La validation de schéma ne détecte aucune violation technique. Cela signifie que les données respectent les contraintes structurelles définies à partir du fichier réel.

Les principaux constats sont les suivants :

- l’identifiant est unique ;
- les variables obligatoires sont présentes ;
- les catégories observées sont cohérentes avec le contenu du dataset ;
- l’âge se situe dans une plage réaliste ;
- les montants de remboursement sont positifs ou nuls ;
- la variable cible est bien binaire.

Cette étape confirme que le jeu de données est techniquement exploitable pour poursuivre l’analyse.

## Analyse de distribution des variables numériques

L’analyse de distribution a pour objectif de mieux comprendre la forme des variables numériques et d’identifier d’éventuelles valeurs aberrantes.

Cette étape est importante pour plusieurs raisons :

- elle permet d’observer les asymétries ou fortes dispersions ;
- elle aide à repérer des valeurs extrêmes ;
- elle prépare les futures étapes de transformation ou de modélisation.

Dans ce projet, les valeurs aberrantes seront détectées à l’aide de deux méthodes statistiques complémentaires :

- la méthode de l’**IQR** ;
- la méthode du **Z-score**.

```python
numeric_analysis_cols = ["LIMIT_BAL", "AGE"] + pay_status_cols + bill_cols + pay_amt_cols

desc_stats = df[numeric_analysis_cols].describe().T
desc_stats["median"] = df[numeric_analysis_cols].median()
desc_stats["skew"] = df[numeric_analysis_cols].skew()
desc_stats["kurtosis"] = df[numeric_analysis_cols].kurtosis()

display(desc_stats)
```

## Lecture synthétique des statistiques descriptives

Les statistiques descriptives montrent que le jeu de données est globalement cohérent, mais que plusieurs variables financières présentent une forte asymétrie et une dispersion importante.

- `AGE` reste dans une plage réaliste, entre 21 et 79 ans.
- `LIMIT_BAL` est asymétrique à droite, avec quelques clients ayant des limites de crédit très élevées.
- Les variables `PAY_*` indiquent que la majorité des clients ont un statut de paiement proche de la normale, mais que certains présentent des retards importants.
- Les variables `BILL_AMT*` et `PAY_AMT*` sont fortement asymétriques, avec des maxima très élevés et des valeurs extrêmes nombreuses.
- Les coefficients élevés de `skew` et de `kurtosis` confirment la présence probable d’outliers, surtout pour les montants facturés et remboursés.

Ces constats justifient pleinement l’étape suivante de détection des valeurs aberrantes à l’aide des méthodes IQR et Z-score.
